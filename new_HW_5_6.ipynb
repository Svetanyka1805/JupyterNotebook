Задание 1
Импортируйте библиотеки pandas и numpy.
Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn. Создайте датафреймы X и y из этих данных.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from sklearn.datasets import load_boston
boston = load_boston()
boston.keys()
```

```python
x = pd.DataFrame(boston.data, columns = boston.feature_names)
x.head(3)
```

```python
y = pd.DataFrame(boston.target, columns = ['Price'])
y.head(3)
```

Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки
составлял 30% от всех данных, при этом аргумент random_state должен быть равен 42.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
x_train, x_test, y_train, y_test
```

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.
Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.
Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
#y_pred.shape
y_pred[:5]
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_train_scaled = pd.DataFrame(x_train_scaled, columns= boston.feature_names)
x_test_scaled = scaler.transform(x_test)
x_test_scaled = pd.DataFrame(x_test_scaled, columns=boston.feature_names)
lr.fit(x_train_scaled, y_train)
y_pred_scaled = lr.predict(x_test_scaled)
r2_core_1 = r2_score(y_test, y_pred_scaled)
r2_core_1
```

```python
from matplotlib import pyplot as plt

feat_weights = pd.Series(lr.coef_.flatten(), index= boston.feature_names)
feat_weights.sort_values(ascending=False).plot(kind='barh')

plt.xlabel("Вес признака")
plt.ylabel("Признак");
```

Разбейте X и y на тренировочный и тестовый наборы данных при помощи функции train_test_split,

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100, stratify=y)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train.info()
#x_test.info()
#y_train.info()
#y_test.info()
```

```python
parameters = [{'n_estimators': [10, 15], 
               'max_features': np.arange(3, 5),
               'max_depth': np.arange(4, 7)}]
parameters
```

```python
grid_srch = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
grid_srch
```

Обучите модель на тренировочном наборе данных (может занять несколько минут).
Просмотрите параметры лучшей модели с помощью атрибута best_params_.
Предскажите вероятности классов с помощью полученнной модели и метода predict_proba.
Из полученного результата (массив Numpy) выберите столбец с индексом 1 (вероятность класса 1) и запишите в массив y_pred_proba. 

```python
grid_srch.fit(x_train, y_train)
```

```python
means = grid_srch.cv_results_['mean_test_score']
stds = grid_srch.cv_results_['std_test_score']


for mean, std, params in zip(means, stds, grid_srch.cv_results_['params']):
        print("%0.3f (+/-%0.03f) for %r" % (mean, std * 2, params))

grid_srch.best_params_
```

```python
y_pred_proba = grid_srch.predict_proba(x_test)[:, 1]
y_pred_proba
```

Из модуля sklearn.metrics импортируйте метрику roc_auc_score.
Вычислите AUC на тестовых данных и сравните с результатом,полученным на тренировочных данных, используя в качестве аргументов массивы y_test и y_pred_proba.

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred_proba)
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba, pos_label=1)

plt.rcParams['figure.figsize'] = 5, 5

plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], color='grey', linestyle='dashed')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC curve');
```
